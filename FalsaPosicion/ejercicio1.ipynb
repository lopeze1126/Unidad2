# El método de la falsa posición


Resolver la ecuación  usando el método de la falsa posición.

```python
import numpy as np

#La función a resolver
P=10.0
T= 353.2
R=0.08205
a=3.599
b=0.04267

def f(v):
    value=P*(v**3) - P*b*(v**2) + a*v- a*b -R*T*(v**2)
    return value
```

```python
# Inputs
p_prev_2=-5.0
p_prev_1=5.0
tol=10**(-15)
error=1.0
Nmax=1000
iteracion=0

print(f(p_prev_2)*f(p_prev_1))
```

```python
#Implementación del algoritmo


while (error>tol):
    p=p_prev_1 - (p_prev_1-p_prev_2)*f(p_prev_1)/((f(p_prev_1))-f(p_prev_2))
    error=abs(p-p_prev_1)
    if (f(p)*f(p_prev_1)<0):
        p_prev_2=p_prev_1
        p_prev_1=p
       
        print("p_n=%.16f\t  error= %.16f\t abs(f(p_n))=%.16f"%(p,error,abs(f(p))))
    elif(f(p)*f(p_prev_1)>0):
        p_prev_1=p_prev_2
        p_prev_2=p
        print("p_n=%.16f\t  error= %.16f\t abs(f(p_n))=%.16f"%(p,error,abs(f(p))))
    iteracion=iteracion+1
  
    if (iteracion>Nmax):
        print ("El método no converge despúes de %d iteraciones"%(Nmax))
        break
    
```
